```python
import yfinance as yf

# Define the ticker symbol for gold priced in USD
gold_ticker = "GC=F"  # This is the futures contract symbol for gold priced in USD on Yahoo Finance

# Download historical market data for gold priced in USD
gold_data = yf.download(gold_ticker, start="2015-01-01", end="2024-04-29")

# Display the first few rows of the data
print(gold_data.head())
```

```python
from forex_python.converter import CurrencyRates
import pandas as pd

# Initialize CurrencyRates object
c = CurrencyRates()

# Define a function to convert USD to INR
def usd_to_inr(date):
    return c.get_rate('USD', 'INR', date)

# Convert gold prices from USD to INR
gold_data['Gold_Price_INR'] = gold_data['Close'] * gold_data.index.map(usd_to_inr)

# Display the converted data
print(gold_data.head())
```
