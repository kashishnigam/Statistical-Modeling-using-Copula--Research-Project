```python
time_series_columns = ['Close_nifty', 'Brent_INR']

def adf_test(timeseries):
    result = adfuller(timeseries, autolag='AIC')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'   {key}: {value}')

# Loop over each column and perform ADF test
for column in time_series_columns:
    print(f"ADF Test Results for {column}:")
    adf_test(df[column])
    print()
```

#### Log Returns

```python
df['Log_Return_Close_nifty'] = np.log(df['Close_nifty'] / df['Close_nifty'].shift(1))
df['Log_Return_Brent_INR'] = np.log(df['Brent_INR'] / df['Brent_INR'].shift(1))
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df = df.dropna()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller, kpss
# Augmented Dickey-Fuller Test
def adf_test(series, title=''):
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series, autolag='AIC')
    labels = ['ADF Statistic', 'p-value', '#Lags Used', 'Number of Observations Used']
    out = pd.Series(result[:4], index=labels)
    for key, value in result[4].items():
        out[f'Critical Value ({key})'] = value
    print(out)
    print('')

# KPSS Test
def kpss_test(series, title=''):
    print(f'KPSS Test: {title}')
    result = kpss(series, regression='c', nlags="auto")
    labels = ['KPSS Statistic', 'p-value', 'Lags Used']
    out = pd.Series(result[:3], index=labels)
    for key, value in result[3].items():
        out[f'Critical Value ({key})'] = value
    print(out)
    print('')
```

```python
# Perform ADF test for Close_nifty log returns
adf_test(df['Log_Return_Close_nifty'], title='Log Return Close_nifty')

# Perform KPSS test for Close_nifty log returns
kpss_test(df['Log_Return_Close_nifty'], title='Log Return Close_nifty')

# Perform ADF test for Brent_INR log returns
adf_test(df['Log_Return_Brent_INR'], title='Log Return Brent_INR')

# Perform KPSS test for Brent_INR log returns
kpss_test(df['Log_Return_Brent_INR'], title='Log Return Brent_INR')
```

```python
# The results of the Augmented Dickey-Fuller (ADF) test and the KPSS test for both `Log_Return_Close_nifty` and `Log_Return_Brent_INR` indicate that the log return series are stationary. Here's how to interpret the results in detail:

# ### Log Return Close_nifty

# #### Augmented Dickey-Fuller (ADF) Test
# - **ADF Statistic**: -9.942944
# - **p-value**: 2.634810e-17 (essentially zero)
# - **Critical Values**:
#   - 1%: -3.432829
#   - 5%: -2.862635
#   - 10%: -2.567353

# **Interpretation**:
# - The ADF statistic (-9.942944) is much lower than any of the critical values.
# - The p-value is significantly lower than 0.05.
# - This means we can reject the null hypothesis that the series has a unit root (is non-stationary).
# - Therefore, the log return series for `Close_nifty` is stationary.

# #### KPSS Test
# - **KPSS Statistic**: 0.085977
# - **p-value**: 0.100000
# - **Critical Values**:
#   - 10%: 0.347
#   - 5%: 0.463
#   - 2.5%: 0.574
#   - 1%: 0.739

# **Interpretation**:
# - The KPSS statistic (0.085977) is much lower than any of the critical values.
# - The p-value is 0.1, which is the highest p-value reported by the KPSS test (indicating a strong case for stationarity).
# - This means we cannot reject the null hypothesis that the series is stationary.
# - Therefore, the log return series for `Close_nifty` is stationary according to the KPSS test as well.

# ### Log Return Brent_INR

# #### Augmented Dickey-Fuller (ADF) Test
# - **ADF Statistic**: -8.825782
# - **p-value**: 1.825269e-14 (essentially zero)
# - **Critical Values**:
#   - 1%: -3.432834
#   - 5%: -2.862638
#   - 10%: -2.567354

# **Interpretation**:
# - The ADF statistic (-8.825782) is much lower than any of the critical values.
# - The p-value is significantly lower than 0.05.
# - This means we can reject the null hypothesis that the series has a unit root (is non-stationary).
# - Therefore, the log return series for `Brent_INR` is stationary.

# **Note**: The warning about the interpolation in the ADF test indicates that the ADF statistic is so low that the corresponding p-value is smaller than those available in the lookup table. The result still strongly suggests stationarity.

# #### KPSS Test
# - **KPSS Statistic**: 0.041981
# - **p-value**: 0.100000
# - **Critical Values**:
#   - 10%: 0.347
#   - 5%: 0.463
#   - 2.5%: 0.574
#   - 1%: 0.739

# **Interpretation**:
# - The KPSS statistic (0.041981) is much lower than any of the critical values.
# - The p-value is 0.1, which is the highest p-value reported by the KPSS test (indicating a strong case for stationarity).
# - This means we cannot reject the null hypothesis that the series is stationary.
# - Therefore, the log return series for `Brent_INR` is stationary according to the KPSS test as well.

# ### Conclusion
# Both the ADF and KPSS tests indicate that the log return series for `Close_nifty` and `Brent_INR` are stationary. This means you can proceed with further analysis, such as modeling with a GARCH-GJR model, without needing additional transformations for stationarity.
```

```python
from arch import arch_model

# Specify the GARCH-GJR model for Close_nifty log returns
model_close_nifty = arch_model(df['Log_Return_Close_nifty'], vol='Garch', p=1, o=1, q=1)

# Specify the GARCH-GJR model for Brent_INR log returns
model_brent_inr = arch_model(df['Log_Return_Brent_INR'], vol='Garch', p=1, o=1, q=1)
```

```python
# Fit the GARCH-GJR model for Close_nifty log returns
results_close_nifty = model_close_nifty.fit(disp='off')
print(results_close_nifty.summary())

# Fit the GARCH-GJR model for Brent_INR log returns
results_brent_inr = model_brent_inr.fit(disp='off')
print(results_brent_inr.summary())
```

```python
# Rescale the log returns
df['Log_Return_Close_nifty_scaled'] = df['Log_Return_Close_nifty'] * 100
df['Log_Return_Brent_INR_scaled'] = df['Log_Return_Brent_INR'] * 100
```

```python
# Refit the models with the rescaled data
model_close_nifty_scaled = arch_model(df['Log_Return_Close_nifty_scaled'], vol='Garch', p=1, o=1, q=1)
results_close_nifty_scaled = model_close_nifty_scaled.fit(disp='off')
print(results_close_nifty_scaled.summary())
```

```python
model_brent_inr_scaled = arch_model(df['Log_Return_Brent_INR_scaled'], vol='Garch', p=1, o=1, q=1)
results_brent_inr_scaled = model_brent_inr_scaled.fit(disp='off')
print(results_brent_inr_scaled.summary())
```

```python
import matplotlib.pyplot as plt
from scipy.stats import rankdata, genpareto
from statsmodels.nonparametric.kde import KDEUnivariate
from statsmodels.stats.diagnostic import acorr_ljungbox
# Plot the standardized residuals
plt.figure(figsize=(10, 6))
plt.plot(results_close_nifty_scaled.resid / results_close_nifty_scaled.conditional_volatility)
plt.title('Standardized Residuals - Close Nifty')
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(results_brent_inr_scaled.resid / results_brent_inr_scaled.conditional_volatility)
plt.title('Standardized Residuals - Brent INR')
plt.show()

# # Perform diagnostic tests
# print("Ljung-Box test for Close Nifty:", results_close_nifty_scaled.test_serial_correlation())
# print("Ljung-Box test for Brent INR:", results_brent_inr_scaled.test_serial_correlation())

# print("ARCH test for Close Nifty:", results_close_nifty_scaled.test_heteroskedasticity())
# print("ARCH test for Brent INR:", results_brent_inr_scaled.test_heteroskedasticity())
```

```python
# Assuming `results` is the fitted GJR-GARCH model for Close_nifty
standardized_residuals_nifty = results_close_nifty_scaled.resid / results_close_nifty_scaled.conditional_volatility

# Assuming `results_brent` is the fitted GJR-GARCH model for Brent_INR
standardized_residuals_brent = results_brent_inr_scaled.resid / results_brent_inr_scaled.conditional_volatility

```

```python
# Step 2: Kernel smoothing for the main body of the distribution
kde_nifty = KDEUnivariate(standardized_residuals['Nifty'])
kde_nifty.fit(kernel='gau', bw='scott')
kde_brent = KDEUnivariate(standardized_residuals['Brent'])
kde_brent.fit(kernel='gau', bw='scott')
```

```python
# Step 3: EVT for tails
threshold_nifty_lower = np.percentile(standardized_residuals['Nifty'], 10)
threshold_nifty_upper = np.percentile(standardized_residuals['Nifty'], 90)
threshold_brent_lower = np.percentile(standardized_residuals['Brent'], 10)
threshold_brent_upper = np.percentile(standardized_residuals['Brent'], 90)
```

```python
# Fit GPD to the tails
params_nifty_lower = genpareto.fit(standardized_residuals['Nifty'][standardized_residuals['Nifty'] < threshold_nifty_lower])
params_nifty_upper = genpareto.fit(standardized_residuals['Nifty'][standardized_residuals['Nifty'] > threshold_nifty_upper])
params_brent_lower = genpareto.fit(standardized_residuals['Brent'][standardized_residuals['Brent'] < threshold_brent_lower])
params_brent_upper = genpareto.fit(standardized_residuals['Brent'][standardized_residuals['Brent'] > threshold_brent_upper])
```
